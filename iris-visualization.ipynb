# The Iris Unfolded: Multidimensional Data Visualization of the Iris Dataset

## Introduction to the Iris Dataset Visualization Project

In this project, we delve into the exploration of the classic Iris dataset, a renowned dataset in the field of data science and machine learning. Comprising 150 samples from three species of Iris flowers (Iris setosa, Iris virginica, and Iris versicolor), the dataset includes four features: sepal length, sepal width, petal length, and petal width, all measured in centimeters. My objective is to employ various data visualization techniques to uncover insights into the dataset. By utilizing a range of plots – scatter, pair, box, violin, histograms, heatmaps, and swarm plots – this analysis aims not only to provide a comprehensive overview of the dataset's characteristics but also to demonstrate the effectiveness of different visualization methods in revealing patterns, distributions, and relationships within the data. This exploration serves as a practice project to enhance my skills in data visualization using Python libraries such as Matplotlib, Seaborn and Pandas, while also offering a deeper understanding of the Iris dataset's multifaceted nature.

## Import libraries for data visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
```

## Acquire dataset to a dataframe

```python
df = pd.read_csv('iris2.csv')
```

## Find appropriate columns for the dataframe

```python
print(df.columns)
```

## Scatter Plot
### This plot is used to explore the relationship between two variables: in this case, sepal length and sepal width.

```python
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='class')
plt.title('Sepal Length vs. Sepal Width by Class')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Iris Species')
plt.show()
```

## Pair Plot
### This plot is used for visualizing the relationships between all pairs of variables at once

```python
warnings.filterwarnings('ignore', category=UserWarning)
sns.pairplot(df, hue='class')
plt.show()
```

## Box Plot
### This plot is used for visualizing the statistical summaries of various features.

```python
feature_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for feature in feature_columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='class', y=feature, data=df)
    plt.title(f'Box plot of {feature} by class')
    plt.show()
```

## Violin Plot
### This plot is used for comparing the distribution of measurements across different species.

```python
sns.violinplot(data=df, x='class', y='sepal_width')
plt.title('Violin Plot of Sepal Width by Class')
plt.xlabel('Iris Species')
plt.ylabel('Sepal Width')
plt.show()
```

## Histograms
### Histograms are used for each numeric feature in the dataset to visualize the frequency distribution:

```python
feature_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for feature in feature_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[feature], kde=True)  # kde=True will plot the kernel density estimate as well
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

## Heatmap of Correlation Matrix
### The heatmap is used to visualize the correlation between different measurements

```python
corr = df.drop('class', axis=1).corr()  # Drop the 'class' column for correlation calculation
sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap of Iris Dimensions')
plt.show()
```

# Swarm Plot
### This plot is used to display all individual measurements without overlap

```python
for feature in feature_columns:
    plt.figure(figsize=(8, 4))
    sns.swarmplot(x='class', y=feature, data=df)
    plt.title(f'Swarm plot of {feature} by class')
    plt.show()
```

# Which plot is the best for visualizing the Iris dataset?

It depends!

Scatter Plot: To explore the relationship between two variables, such as sepal length and sepal width. Scatter plots are excellent for visualizing how two measurements correlate with each other and can be enhanced by using color to represent different species.

Pair Plot (Pairwise Scatter Plot): Since the Iris dataset has multiple measurements, a pair plot can help visualize the relationships between all pairs of variables at once. It's especially useful for spotting correlations and clusters across multiple dimensions.

Box Plot: For visualizing the statistical summaries of various features. Box plots show the median, quartiles, and outliers for each measurement, which can be particularly useful for seeing how different iris species differ in terms of petal and sepal dimensions.

Violin Plot: These are similar to box plots but also include a kernel density estimate. They are useful for comparing the distribution of measurements across different species.

Histogram: To look at the frequency distribution of a single variable, such as petal length. It's a simple way to see the distribution and central tendency of measurements.

Heatmap of Correlation Matrix: To visualize the correlation between different measurements. This can be useful to quickly identify which dimensions have the strongest relationships.

Swarm Plot: This can be a good choice if you want to display all individual measurements without overlap, which can sometimes occur in scatter plots.
